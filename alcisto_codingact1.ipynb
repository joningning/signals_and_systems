```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def x(t):
    return 9* np.cos(200*np.pi*t +0.4*np.pi)

timeMin =-0.02
timeMax = 0.02

sampleIntervals=[0.0075, 0.005, 0.0005]
```

```python
fig, axes=plt.subplots(3,1,figsize=(10,15))

for i, ts in enumerate(sampleIntervals):
    t_samples = np.arange(timeMin, timeMax, ts)
    x_samples = x(t_samples)
    axes[i].plot(t_samples, x_samples, marker='o', linestyle='-')
    axes[i].set_title(f'Ts = {ts} s')
    axes[i].set_xlabel('Time (s)')
    axes[i].set_ylabel('x(t)')
    axes[i].grid(True)
```
